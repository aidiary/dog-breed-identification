## Split to Train/Valid

```python
from sklearn.model_selection import train_test_split
```

```python
ids = train_labels['id'].values
labels = train_labels['breed'].values
```

```python
ids
```

```python
labels
```

```python
train_id, valid_id, train_label, valid_label = train_test_split(ids, labels, test_size=0.1, random_state=42, stratify=labels)
```

```python
train_id.shape, train_label.shape, valid_id.shape, valid_label.shape
```

```python
from collections import Counter
print(Counter(train_label))
print(Counter(valid_label))
```

```python
len(set(train_label)), len(set(valid_label))
```

```python
import shutil
for n in valid_id:
    src = '../data/train/{}.jpg'.format(n)
    tgt = '../data/valid/{}.jpg'.format(n)
    shutil.move(src, tgt)
```

```python
# クラスごとにディレクトリを作成
```

```python
classes = set(train_label)
classes
```

```python
len(classes)
```

```python
import os
for cls in classes:
    os.makedirs('../data/train/{}'.format(cls), exist_ok=True)
    os.makedirs('../data/valid/{}'.format(cls), exist_ok=True)
```

```python
import shutil
for n, cls in zip(train_id, train_label):
    src = '../data/train/{}.jpg'.format(n)
    tgt = '../data/train/{}/{}.jpg'.format(cls, n)
    shutil.move(src, tgt)
```

```python
for n, cls in zip(valid_id, valid_label):
    src = '../data/valid/{}.jpg'.format(n)
    tgt = '../data/valid/{}/{}.jpg'.format(cls, n)
    shutil.move(src, tgt)
```
